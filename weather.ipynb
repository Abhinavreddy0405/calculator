```python
import random
import time

class Weather:
    def __init__(self):
        self.temperature = 0
        self.humidity = 0
        self.pressure = 0

    def update(self):
        self.temperature = random.randint(20, 40)
        self.humidity = random.randint(30, 90)
        self.pressure = random.randint(980, 1050)

    def display(self):
        print("Temperature:", self.temperature, "°C")
        print("Humidity:", self.humidity, "%")
        print("Pressure:", self.pressure, "hPa")
        print("-" * 25)

weather = Weather()

for i in range(5):
    weather.update()
    weather.display()
    time.sleep(2)
```
